```python
import pandas as pd
df = pd.read_csv("./Toy_Data.csv")
df
```

```python
# Convert a column to datetime
df['date_column'] = pd.to_datetime(df['date'])
df
```

```python
df['year'] = df['date_column'].dt.year
df['month'] = df['date_column'].dt.month
df['day'] = df['date_column'].dt.day
df['weekday'] = df['date_column'].dt.weekday  # Monday = 0, Sunday = 6
df['quarter'] = df['date_column'].dt.quarter
df
```

```python
df = df.sort_values(['date_column','year','weekday']).reset_index(drop=True)
df['diff_days'] = df.groupby(['year','weekday'])['date_column'].diff().dt.days
df
```

```python
df['date_plus_10_days'] = df['date_column'] + pd.Timedelta(days=10)
df
```

```python
df['date_diff'] = df['date_plus_10_days'] - df['date_column']
df
```

```python
df['diff_days'].isna().sum()
```

```python
df['diff_days'] = df['diff_days'].fillna(pd.Timestamp('2020-01-01')) # df['diff_days'].fill_na(pd.Timestamp('2020-01-01'), inplace=True)
df
```

```python
df['previous_day'] = df['value'].shift(1)  # Shift by 1 row (previous day)
df['next_day'] = df['value'].shift(-1)  # Shift forward by 1 row (next day)
df
```

```python
import pandas as pd

# Get the current date
current_date = pd.Timestamp.today()

# Subtract 1 quarter (3 months) from the current date
one_quarter_ago = current_date - pd.DateOffset(months=3)

# Example to filter data from a DataFrame `df` with a 'date' column
# assuming the 'date' column is of type datetime
filtered_df = df[df['date'] >= one_quarter_ago]

# Display the date range
print("Current Date:", current_date)
print("Date 1 Quarter Ago:", one_quarter_ago)
```
